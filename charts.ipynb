```python
import re
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
res_time = []
res_ram = []
for name in ['res_not_opt-docker1.txt', 'res_opt-docker1.txt', 'res_not_opt-docker3.txt', 'res_opt-docker3.txt']:
    with open(f"./results/{name}", "r") as f:
        for line in f.readlines():
            numbers = re.findall(r'\d+.\d+', line)
            res_time.append(float(numbers[0]))
            res_ram.append(float(numbers[1]))
```

```python
len(res_time)
```

```python
len(res_ram)
```

```python
fig, axes = plt.subplots(ncols=4, figsize=(20,5))
for col in range(4):
    sns.histplot(res_time[(col)*30:(col+1)*30], kde=True, stat="density", bins=10, color='red', ax=axes[col])
    if col == 0:
        axes[col].set_title("  \n res_not_opt-docker1.txt")
    if col ==1:
        axes[col].set_title("  \n res_opt-docker1.txt")    
    if col ==2:
        axes[col].set_title("  \n res_not_opt-docker3.txt")    
    if col ==3:
        axes[col].set_title("  \n res_opt-docker3.txt")
    axes[col].set_xlabel(" ")
    axes[col].set_ylabel("")
    axes[col].legend()
plt.show()
```

```python
fig, axes = plt.subplots(ncols=4, figsize=(20,5))
for col in range(4):
    sns.histplot(res_ram[(col)*30:(col+1)*30], kde=True, stat="density", bins=10, color='red', ax=axes[col])
    if col == 0:
        axes[col].set_title(" RAM \n res_not_opt-docker1.txt")
    if col ==1:
        axes[col].set_title(" RAM \n res_opt-docker1.txt")    
    if col ==2:
        axes[col].set_title(" RAM \n res_not_opt-docker3.txt")    
    if col ==3:
        axes[col].set_title(" RAM \n res_opt-docker3.txt")
    axes[col].set_xlabel(" RAM")
    axes[col].set_ylabel("")
    axes[col].legend()
plt.show()
```
